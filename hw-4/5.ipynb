```python
import numpy as np
import matplotlib.pyplot as plt

errors: list = []

n = 100  # Размер матрицы
A = np.random.rand(n, n)  # Генерация случайной матрицы

U, s, VT = np.linalg.svd(A)
Sigma = np.diag(s)

for r in range(2, n + 1):
    A_approx = U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]
    error = np.linalg.norm(A - A_approx, 'fro')
    errors.append(error)

plt.plot(range(2, n + 1), errors, marker='o')
plt.title('Зависимость ошибки аппроксимации от ранга r')
plt.xlabel('Ранг r')
plt.ylabel('Ошибка аппроксимации')
plt.show()
```
